### Representação Binária e Soma Infinita
   - Considere um número \$y \in (0, 1)\$. Esse número pode ser escrito em expansão binária como $ y = (0.y_1 y_2 y_3 \ldots)_2 $ onde cada $ y_i \in \{0, 1\} $. A soma $ X = \sum_{i=1}^{\infty} 2^{-i} X_i $ representa exatamente essa expansão binária, onde $ X_i $ é a $ i $-ésima posição na expansão binária e $X = y$ . 

### Comparação Lexicográfica
   - Se \$ 0.X_1 X_2 X_3 \ldots \$ e \$ 0.x_1 x_2 x_3 \ldots \$ são duas sequências binárias que representam os números decimais \$ X \$ e \$ x \$, respectivamente, então:

        \$ 0.X_1 X_2 X_3 \ldots < 0.x_1 x_2 x_3 \ldots \$ se, e somente se, \$ X < x \$ em sua forma decimal.
   - Considere duas sequências binárias, \$ 0.X_1 X_2 X_3 \ldots \$ e \$ 0.x_1 x_2 x_3 \ldots \$, então \$ 0.X_1 X_2 X_3 \ldots < 0.x_1 x_2 x_3 \ldots \$ se a primeira sequência contém o primeiro digito que as difere menor que a segunda sequência.
   
       Exemplo: (0.101) < (0.110), pois elas se diferem no 2º dígito e 0 < 1.

### Justificação da Probabilidade

$$
\begin{aligned}
P(X < x) &= P(\sum_{i=1}^{\infty} 2^{-i} X_i < x) \\
            &= P(X_1, X_2, \ldots < x_1, x_2, \ldots) \\
            &= P(\{X_1 < x_1\} \cup \{X_1 = x_1, X_2 < x_2\} \cup \ldots) \\
            &= P(X_1 < x_1) + P(X_1 = x_1, X_2 < x_2) +\ldots\\
            &= \sum_{i=1}^{\infty} P(X_i < x_i) P(X_1 = x_1)^{i-1} \\
            &= \sum_{i=1}^{\infty} P(X_i < x_i) \left(\frac{1}{2}\right) ^{i-1} \\
            &= 2 \sum_{i=1}^{\infty} P(X_i < x_i) 2^{-i} \\
            &= 2 \sum_{i=1}^{\infty} x_i 2^{-1} 2^{-i} \\
            &= x
\end{aligned}
$$


Utilizamos na 4ª igualdade o axioma 1 da função $P()$, visto que os eventos são disjuntos. 

Já na 5ª, usamos que os $X_i$ são i.i.d, segue assim que  $P(X_1 = x_1) = P(X_j = x_j), \ \forall j \in (1,\infty)$, logo:

$$P(X_1 = x_1, \ldots, X_{i-1} = x_{i-1}, X_i < x_i) = P(X_i < x_i) P(X_1 = x_1)^{i-1} \forall i \in (1,\infty)$$

Perceba também que $P(X_1 = x_1) = P(\{X_1 = 0, x_1 = 0\} \cup \{X_1 = 1, x_1 = 1\}) = \frac{1}{2}$.

Por fim, utilizamos que $P(X_i < x_i) = \frac{1}{2}$ se $x_i = 1$ e $P(X_i < x_i) = 0$ se $x_i = 0$. Assim, podemos escrever que $P(X_i < x_i) = \frac{x_i}{2} = x_i 2^{-1}$.

### Precisão de Ponto Flutuante

Computadores representam números usando precisão finita. Para números de ponto flutuante de precisão dupla (double precision), o padrão IEEE 754 define que a precisão é aproximadamente $2^{-53}$. Isso significa que a menor diferença detectável entre dois números distintos é da ordem de $2^{-53}$. Repare que isso varia de máquina para máquina.

### Truncamento da Soma Infinita

A série $X = \sum_{i=1}^{\infty} 2^{-i} X_i$ é uma série infinita, mas na prática, não podemos somar infinitos termos em um computador. Precisamos truncar a soma após um número finito de termos, $k$, e queremos garantir que o erro introduzido por truncar a série seja menor do que a precisão da máquina.

### Justificativa para a Soma dos Termos

Queremos encontrar $k$ tal que a soma dos termos $2^{-i}$ para $i > k$ seja menor do que a menor diferença que a máquina pode detectar, ou seja, $2^{-53}$. Assim, garantimos que a contribuição dos termos ignorados é menor do que a precisão da máquina, tornando o erro de truncamento insignificante.

### Análise da Soma dos Termos

Consideramos a soma dos termos $2^{-i}$ para $i > k$:

$$
\sum_{i=k+1}^{\infty} 2^{-i}
$$

Esta soma é uma série geométrica infinita. A soma de uma série geométrica infinita a partir do termo $i = k+1$ é:

$$
\sum_{i=k+1}^{\infty} 2^{-i} = 2^{-(k+1)} + 2^{-(k+2)} + 2^{-(k+3)} + \ldots
$$

Podemos fatorar $2^{-(k+1)}$:

$$
\sum_{i=k+1}^{\infty} 2^{-i} = 2^{-(k+1)} \left( 1 + 2^{-1} + 2^{-2} + \ldots \right)
$$

A soma da série geométrica dentro dos parênteses é uma série geométrica:

$$
1 + 2^{-1} + 2^{-2} + \ldots = \frac{1}{1 - 2^{-1}} = 2
$$

Portanto:

$$
\sum_{i=k+1}^{\infty} 2^{-i} = 2 \cdot 2^{-(k+1)} = 2^{-k}
$$

Queremos que essa soma seja menor que a precisão da máquina $2^{-53}$:

$$
2^{-k} < 2^{-53}
$$

Tomando o logaritmo base 2 de ambos os lados, obtemos:

$$
-k < -53 \implies k > 53
$$

### Conclusão

Para garantir que a contribuição dos termos truncados seja menor do que a precisão da máquina, devemos escolher $k$ tal que:

$$
k > 53
$$

Portanto, truncar a série após $k = 54$ termos é suficiente para alcançar a precisão numérica da máquina.


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Parte (b): Gerador de números uniformemente distribuídos usando truncamento
def uniform_generator(k):
    # usando o fato que randint(0,2) gera 0 ou 1 com mesma probabilidade
    X = sum(2**-i * np.random.randint(0, 2) for i in range(1, k+1))
    return X
```

```python
# Encontrando o valor de k
machine_precision = np.finfo(float).eps
k = int(np.ceil(-np.log2(machine_precision)))
print("Valor da precisão k: ", k)

# Parte (c): Amostragem de n = 1000 variáveis uniformes e plotagem do histograma
n = 1000
samples = [uniform_generator(k) for _ in range(n)]

plt.figure(figsize=(8, 4))
plt.hist(samples, bins=18, density=True, alpha=0.65, edgecolor='black')
plt.title('Histograma de 1000 variáveis uniformemente distribuídas em [0, 1]')
plt.xlabel('Valor')
plt.ylabel('Frequência')
plt.show()
```

### Universalidade da Uniforme
Sabemos que, utilizando a universalidade da distribuição uniforme, se \$U \sim \text{Unif}(0,1)\$, então a v.a \$Y=F_X^{-1}(U)\$ tem CDF \$F_X\$.

### Inversa da CDF da Exponencial
Queremos encontrar a inversa da CDF, \$F_X^{-1}(u)\$, onde \$u \in (0, 1)\$ e \$X \sim \text{Expo}(\lambda)\$, para podermos gerar uma exponencial a partir de uma unifrome. Portanto, temos:

$$
\begin{aligned}
u &= F_X(x) \\
u &= 1 - e^{-\lambda x} \\
e^{-\lambda x} &= 1 - u \\
-\lambda x &= \ln(1 - u) \\
x &= -\frac{1}{\lambda} \ln(1 - u) \\
\end{aligned}
$$

Portanto, a inversa da CDF da distribuição exponencial é:

$$F_X^{-1}(u) = -\frac{1}{\lambda} \ln(1 - u)$$


```python
def expo_generator(u, lambd=1):
    return -np.log(1 - u) / lambd
```

```python
n = 1000
uniform_samples = np.array([uniform_generator(k) for _ in range(n)]) # gerando uma Unif(0,1) para gerar a Expo(1)
expo_samples = expo_generator(uniform_samples)

plt.figure(figsize=(8, 4))
plt.hist(expo_samples, bins=18, density=True, alpha=0.65, edgecolor='black')
plt.title('Histograma de 1000 variáveis distribuídas por uma Expo(1)')
plt.xlabel('Valor')
plt.ylabel('Frequência')
plt.show()
```

### Box-Muller

Seja $T \sim \text{Expo}(1) \text{ e } U \sim \text{Unif}(0, 2\pi)$  independentes. Defina $X = \sqrt{2T} \cos(U)$, então $X \sim \mathcal{N}(0,1)$.


```python
# Box Muller
def normal_generator(t,u):
    # t Expo(1) e u Unif(0,2pi)
    x = np.sqrt(2*t)*np.cos(u)
    return x
```

```python
uniform_samples = np.array([uniform_generator(k) for _ in range(n)]) # gerando uma Unif(0,1) para gerar a Expo(1)
T = expo_generator(uniform_samples)
U = np.array([uniform_generator(k) for _ in range(n)]) * 2 * np.pi # gerando Unif(0, 2pi) independente da exponencial

normal_samples = normal_generator(T,U)

plt.figure(figsize=(8, 4))
plt.hist(normal_samples, bins=18, density=True, alpha=0.65, edgecolor='black')
plt.title('Histograma de 1000 variáveis distribuídas por uma Normal(0,1)')
plt.xlabel('Valor')
plt.ylabel('Frequência')
plt.show()
```
